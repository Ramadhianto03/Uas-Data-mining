```python
data2 = data.copy()
```

```python
#rata-rata
avg_sex = data2['sex'].mean()
avg_cp = data2['cp'].mean()
avg_chol = data2['chol'].mean()
avg_fbs = data2['fbs'].mean()
avg_restecg = data2['restecg'].mean()
avg_exng = data2['exng'].mean()
avg_oldpeak = data2['oldpeak'].mean()
avg_slp = data2['slp'].mean()
avg_caa = data2['caa'].mean()
avg_output = data2['output'].mean()
```

```python
#now replace avg to columns 
data2['sex'].replace({0:avg_sex},inplace=True)
data2['cp'].replace({0:avg_cp},inplace=True)
data2['chol'].replace({0:avg_chol},inplace=True)
data2['fbs'].replace({0:avg_fbs},inplace=True)
data2['restecg'].replace({0:avg_restecg},inplace=True)
```

```python
print('Banyak baris nilai nol untuk age :{0}'.format(len(data2.loc[data['age']==0])))
print('Banyak baris nilai nol untuk sex :{0}'.format(len(data2.loc[data2['sex']==0])))
print('Banyak baris nilai nol untuk cp :{0}'.format(len(data2.loc[data2['cp']==0])))
print('Banyak baris nilai nol untuk trtbps :{0}'.format(len(data2.loc[data2['chol']==0])))
print('Banyak baris nilai nol untuk chol :{0}'.format(len(data2.loc[data2['fbs']==0])))
print('Banyak baris nilai nol untuk fbs :{0}'.format(len(data2.loc[data2['fbs']==0])))
print('Banyak baris nilai nol untuk restecg :{0}'.format(len(data2.loc[data2['restecg']==0])))
print('Banyak baris nilai nol untuk thalachh :{0}'.format(len(data2.loc[data2['age']==0])))
print('Banyak baris nilai nol untuk exng :{0}'.format(len(data2.loc[data2['exng']==0])))
print('Banyak baris nilai nol untuk oldpeak :{0}'.format(len(data2.loc[data2['oldpeak']==0])))
print('Banyak baris nilai nol untuk slp :{0}'.format(len(data2.loc[data2['slp']==0])))
print('Banyak baris nilai nol untuk caa :{0}'.format(len(data2.loc[data2['caa']==0])))
print('Banyak baris nilai nol untuk output :{0}'.format(len(data2.loc[data2['output']==0])))
```

```python
from sklearn.model_selection import train_test_split

X = data2.drop('output',axis=1)
y = data2['output']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.svm import SVC
SVC_model = SVC()
```

```python
SVC_model.fit(X_train,y_train)
```

```python
svc_prediction = SVC_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,svc_prediction))
```

```python
print(classification_report(y_test,svc_prediction))
```
